## 1.2 Adding a new columnwith an array 

####  dataframe has 1000 rows, so we need to create an array with 1000 elements

```python
# import numpy
import numpy as np
```

```python
# create an array of 1000 elements
language_score = np.arange(0,1000)
```

```python
# length of the array
len(language_score)
```

```python
# adding a new column to a dataframe with an array
df_exams['language score'] = language_score
df_exams
```

```python
# create random integer numbers between 1 and 100
int_language_score = np.random.randint(1,100,size = 1000)
```

```python
# min value inclusive and higher value excliusive
min(int_language_score)
max(int_language_score)
```

```python
# adding a new column to a dataframe with an array
df_exams['language score'] = int_language_score
df_exams
```

```python
# create random float numbers between 1 and 100
f_val = np.random.uniform(1,100,size = 1000)
f_val
```

```python
# rounding float numbers to 2 decimal values
f_val = [round(x,2) for x in f_val]
f_val
```
